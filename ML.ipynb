# 1. Analyse statistique univariée

➤ Valeurs aberrantes

```python
dataset.describe()
```

```python

numeric_cols = dataset.select_dtypes(include='number')


Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1


outliers = ((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR)))


print("Nombre de valeurs aberrantes par colonne :")
print(outliers.sum())
```

```python
numeric_cols = dataset.select_dtypes(include='number')


plt.figure(figsize=(15, 6))
sns.boxplot(data=numeric_cols, orient='h', palette='Set2')
plt.title("Boxplot des variables numériques (détection des valeurs aberrantes)")
plt.xlabel("Valeurs")
plt.show()
```

```python

winsorized = numeric_cols.copy()
for col in numeric_cols.columns:
    Q1 = numeric_cols[col].quantile(0.25)
    Q3 = numeric_cols[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    winsorized[col] = numeric_cols[col].clip(lower=lower, upper=upper)

dataset[numeric_cols.columns] = winsorized
```

```python
from scipy.stats import skew, kurtosis, shapiro, normaltest
quant_vars = [
    col for col in dataset.select_dtypes(include=['float64', 'int64']).columns
    if col not in ['co_id'] 
]
print("Variables quantitatives détectées :", quant_vars)


for var in quant_vars:
    data = dataset[var].dropna()
    
    
    skewness = skew(data)
    kurt = kurtosis(data) 
    
    
    sample_data = data.sample(n=5000, random_state=42) if len(data) > 5000 else data
    shapiro_stat, shapiro_p = shapiro(sample_data)
    
    
    dagostino_stat, dagostino_p = normaltest(data)
    
    print(f"\n📊 Analyse de la forme - {var}")
    print(f"➤ Asymétrie (skewness)     : {skewness:.2f} {'➡ droite' if skewness > 0 else '➡ gauche' if skewness < 0 else 'symétrique'}")
    print(f"➤ Aplatissement (kurtosis) : {kurt:.2f} {'➡ leptokurtique' if kurt > 0 else '➡ platykurtique' if kurt < 0 else 'normal'}")
    print(f"➤ Test de Shapiro-Wilk     : p = {shapiro_p:.4f} {'(non normale)' if shapiro_p < 0.05 else '(normale)'}")
    print(f"➤ Test de D’Agostino       : p = {dagostino_p:.4f} {'(non normale)' if dagostino_p < 0.05 else '(normale)'}")

   
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    sns.histplot(data, kde=True, ax=axes[0], color='skyblue', bins=40)
    axes[0].set_title(f"Histogramme & KDE - {var}")
    
    sns.boxplot(x=data, ax=axes[1], color='lightgreen')
    axes[1].set_title(f"Boxplot - {var}")
    
    plt.suptitle(f"Analyse graphique - {var}")
    plt.tight_layout()
    plt.show()
```

➤ Valeurs aberrantes

```python
cat_cols = dataset.select_dtypes(include='object').columns

for col in cat_cols:
   
    top_modalities = dataset[col].value_counts().nlargest(10).index
    freq_abs = dataset[col].value_counts().loc[top_modalities]
    freq_rel = dataset[col].value_counts(normalize=True).loc[top_modalities] * 100

    print(f"\n📊 {col} - Fréquences absolues (top 10) :\n{freq_abs}")
    print(f"\n🔢 {col} - Fréquences relatives (%) (top 10) :\n{freq_rel.round(2)}")

    plt.figure(figsize=(7, 4))
    ax = sns.countplot(data=dataset[dataset[col].isin(top_modalities)],
                       x=col,
                       order=top_modalities,
                       palette='pastel')
    plt.title(f"Distribution de {col} (Top 10 modalités)")
    plt.xlabel(col)
    plt.ylabel("Nombre d'observations")

    
    total = len(dataset)
    for p in ax.patches:
        count = p.get_height()
        percentage = f'{100 * count / total:.1f}%'
        ax.annotate(percentage,
                    (p.get_x() + p.get_width() / 2, count),
                    ha='center', va='bottom', fontsize=9)

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# 2. k Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train_sc, y_train)

y_pred_knn = knn.predict(X_test_sc)

print("Comparaison premiers résultats (y_test vs prédiction) :")
for i in range(10):
    print(f"Vérité : {y_test.iloc[i] if hasattr(y_test, 'iloc') else y_test[i]} - Prédiction : {y_pred_knn[i]}")

print("\n--- Évaluation du modèle KNN ---")
print("Matrice de confusion :\n", confusion_matrix(y_test, y_pred_knn))
print(f"Accuracy : {accuracy_score(y_test, y_pred_knn):.4f}")
print(f"Recall : {recall_score(y_test, y_pred_knn):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_knn):.4f}")
print("\nRapport de classification :\n", classification_report(y_test, y_pred_knn))
```

```python
param_grid = {'n_neighbors': [1, 3, 5, 7, 9, 11, 13]}
knn = KNeighborsClassifier()

grid_search = GridSearchCV(knn, param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train_sc, y_train)

print("\n--- Résultats GridSearchCV ---")
print("Meilleur nombre de voisins :", grid_search.best_params_['n_neighbors'])
print(f"Meilleure accuracy en CV : {grid_search.best_score_:.4f}")
```

```python
knn_optimized = grid_search.best_estimator_

y_pred_knn_opt = knn_optimized.predict(X_test_sc)
print("\n--- Évaluation modèle KNN optimisé ---")
print("Matrice de confusion :\n", confusion_matrix(y_test, y_pred_knn_opt))
print(f"Accuracy : {accuracy_score(y_test, y_pred_knn_opt):.4f}")
print(f"Recall : {recall_score(y_test, y_pred_knn_opt):.4f}")
print(f"Precision : {precision_score(y_test, y_pred_knn_opt):.4f}")
print("\nRapport de classification :\n", classification_report(y_test, y_pred_knn_opt))
```

# Partie finale : Évaluation et comparaison de tous les modèles

```python
results = {
    "Modèle": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1-score": []
}

def evaluate_model(name, y_true, y_pred):
    results["Modèle"].append(name)
    results["Accuracy"].append(accuracy_score(y_true, y_pred))
    results["Precision"].append(precision_score(y_true, y_pred))
    results["Recall"].append(recall_score(y_true, y_pred))
    results["F1-score"].append(f1_score(y_true, y_pred))

evaluate_model("KNN", y_test, y_pred_knn)
evaluate_model("KNN optimisé", y_test, y_pred_knn_opt)
evaluate_model("Logistic Regression", y_test, y_pred_LR)
evaluate_model("Random Forest", y_test, y_pred_rf)
evaluate_model("Decision Tree", y_test, y_pred_dt)

df_results = pd.DataFrame(results)

print("\n--- Comparaison des modèles ---\n")
print(df_results.sort_values(by="F1-score", ascending=False).round(4))

```

```python
plt.figure(figsize=(12, 6))
df_results_melted = df_results.melt(id_vars="Modèle", var_name="Métrique", value_name="Score")
sns.barplot(data=df_results_melted, x="Modèle", y="Score", hue="Métrique")
plt.title("Comparaison des performances des modèles")
plt.ylim(0, 1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import f1_score

model_names = ['KNN', 'KNN optimisé', 'Logistic Regression', 'Random Forest', 'Decision Tree']
model_preds = [y_pred_knn, y_pred_knn_opt, y_pred_LR, y_pred_rf, y_pred_dt]

f1_scores = [f1_score(y_test, y_pred) for y_pred in model_preds]

df_f1 = pd.DataFrame({
    'Modèle': model_names,
    'F1-score': f1_scores
}).sort_values(by='F1-score', ascending=False).reset_index(drop=True)

print("\n--- Évaluation finale basée sur le F1-score ---\n")
print(df_f1.round(4))

plt.figure(figsize=(10, 5))
sns.barplot(data=df_f1, x='Modèle', y='F1-score', palette='viridis')
plt.title("Comparaison des modèles selon le F1-score", fontsize=14)
plt.ylabel("F1-score")
plt.ylim(0, 1)
plt.xlabel("")
plt.xticks(rotation=20)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
